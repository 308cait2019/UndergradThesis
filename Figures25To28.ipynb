```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("Zebra_Mussel_Dry_Weights.csv")
df3a = df
print("Graph 25: Zebra mussel dry weights by only flesh")


G3a = nx.Graph()

G3a.add_nodes_from(df3a['flesh'])
G3a.add_nodes_from(df3a['species'])

# Construct the edge_list and add it to graph with weights:
edge_list = df3a[['species','flesh','flesh']] #changed syntax
edge_list2 = list(edge_list.itertuples(index=False))

G3a.add_weighted_edges_from(edge_list2)

print("Total number of nodes: ", int(G3a.number_of_nodes()))
print("Total number of edges: ", int(G3a.number_of_edges()))


# Create dictionary of edge labels: 

#edge_labels_key = df3a[['species','flesh']].itertuples(index=False)
#edge_labels_dict = dict(zip(edge_labels_key, df3a['flesh'].round(decimals = 4)))

# Style and show graph: 
pos = nx.shell_layout(G3a,scale=1)

# Make graph styling 
plt.figure(3,figsize=(12,12))

color_map = ['red' if node == 'species' else 'violet' for node in G3a]        

nx.draw(G3a, pos, node_color = color_map, font_color = 'black', edge_color='grey', width=1, linewidths=1,
    node_size=250,  alpha=0.9, with_labels = False)
        
#nx.draw_networkx_edge_labels(G3a, pos, edge_labels = edge_labels_dict, font_color = 'purple')
plt_size = (105,150)
label_offset = -20
plt.show()
```

```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("Zebra_Mussel_Dry_Weights.csv")
df3b = df
print("Graph 26: Zebra mussel dry weights by only shell")


G3b = nx.Graph()

G3b.add_nodes_from(df3b['species'])
G3b.add_nodes_from(df3b['shell'])

# Construct the edge_list and add it to graph with weights:
edge_list = df3b[['species','shell','shell']] #changed syntax
edge_list2 = list(edge_list.itertuples(index=False))

G3b.add_weighted_edges_from(edge_list2)

print("Total number of nodes: ", int(G3b.number_of_nodes()))
print("Total number of edges: ", int(G3b.number_of_edges()))


# Create dictionary of edge labels: 

edge_labels_key = df3b[['species','shell']].itertuples(index=False)
edge_labels_dict = dict(zip(edge_labels_key, df3b['shell'].round(decimals = 4)))

# Style and show graph: 
pos = nx.shell_layout(G3b,scale=1)

# Make graph styling 
plt.figure(3,figsize=(12,12))

color_map = ['red' if node == 'species' else 'violet' for node in G3b]        

nx.draw(G3b, pos, node_color = color_map, font_color = 'black', edge_color='grey', width=1, linewidths=1,
    node_size=250,  alpha=0.9, with_labels = True)
        
nx.draw_networkx_edge_labels(G3b, pos, edge_labels = edge_labels_dict, font_color = 'purple')
plt_size = (105,150)
label_offset = -20
plt.show()
```
