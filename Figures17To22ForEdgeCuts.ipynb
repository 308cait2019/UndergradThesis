```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("wc1A2014.csv")
df_white14 = df[df['species']== 'white perch'] #changed df_white

print("Graph 1.a.1: white perch in 2014")

#df.iloc[0:344]

G1A = nx.Graph()

G1A.add_nodes_from(df_white14['prey'])
G1A.add_nodes_from(df_white14['species'])

#edge_list = df.loc[0::346, 'prey_weight_g']

# Here the changes begin:


# Construct the edge_list and add it to graph with weights:
threshold_weight = 1.500
keep_prey = []
for prey_name in df_white14['prey'].unique():
    small_df = df_white14[df_white14['prey']==prey_name]
    if sum(small_df['prey_weight_g']) <= threshold_weight:
        keep_prey.append(prey_name)
       
df_edges = df_white14[df_white14['prey'].isin(keep_prey)]

edge_list = df_edges[['species','prey','prey_weight_g']] #changed syntax
edge_list2 = list(edge_list.itertuples(index=False))

G1A.add_weighted_edges_from(edge_list2)

print("Total number of nodes: ", int(G1A.number_of_nodes()))
print("Total number of edges: ", int(G1A.number_of_edges()))


# Create dictionary of edge labels: 


edge_labels_key = df_edges[['species','prey']].itertuples(index=False)
edge_labels_dict = dict(zip(edge_labels_key, df_edges['prey_weight_g'].round(decimals = 4)))


# Style and show graph: 

pos = nx.shell_layout(G1A,scale=1)

# Make graph styling
plt.figure(3,figsize=(12,12))
nx.draw(G1A, pos, edge_color='black', width=1, linewidths=1,
    node_size=250,  alpha=0.9, with_labels = True)
nx.draw_networkx_edge_labels(G1A, pos, edge_labels = edge_labels_dict)
plt_size = (105,150)
label_offset = -20
plt.show()
```

```python
from pyvis.network import Network
# Only run once- if run again without recreating the graph in networkx, all edge weight data disappears 
net = Network(notebook = True)
net.from_nx(G1A)

# Get edge data
edges = net.get_edges()
for edge in edges:
    # add a label to the edge dictionary. 
    edge['label'] = str(round(edge['width'],4)) + " g"
    
    # comment this out after the first time run
    edge['original width'] = edge['width']
    
    # play with scaling the edge weight to scale the thickness of the edges
    edge['width'] = edge['original width'] * 5
    
# change the styling, location etc 
# turn off physics on nodes if want them to stay where dragged to
# change the length of edges by changing size of the gravitational constant - under 'physics' at the bottom
net.show_buttons(filter_ = True)

# show graph!
net.show("white_perch_2014.html") 
```

```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from networkx.algorithms import bipartite 
    
T1 = pd.read_csv("wc1A2014.csv")

print("Figure 18: edge cuts to both perch species in 2014")

#df.iloc[238:522]

G1A = nx.Graph()

G1A.add_nodes_from(T1['prey'], bipartite = 1)
G1A.add_nodes_from(T1['species'], bipartite = 0) 

#edge_list = df.loc[0::523, 'prey_weight_g']

# Here the changes begin:


# Construct the edge_list and add it to graph with weights:
threshold_weight = 1.500
keep_prey_white = []
keep_prey_yellow = []
for prey_name in T1['prey'].unique():
    small_df = T1[T1['prey']==prey_name]
    small_df_white = small_df[small_df['species'] == 'white perch']
    if sum(small_df_white['prey_weight_g']) <= threshold_weight:
        keep_prey_white.append(prey_name)
    small_df_yellow = small_df[small_df['species'] == 'yellow perch']
    if sum(small_df_yellow['prey_weight_g']) <= threshold_weight:
        keep_prey_yellow.append(prey_name)
        
df_edges_yellow = T1[T1['species'] == 'yellow perch']
df_edges_yellow = df_edges_yellow[df_edges_yellow['prey'].isin(keep_prey_yellow)]
df_edges_white = T1[T1['species'] == 'white perch']
df_edges_white = df_edges_white[df_edges_white['prey'].isin(keep_prey_white)]
df_edges = pd.concat([df_edges_yellow, df_edges_white])

edge_list = df_edges[['species','prey','prey_weight_g']]
edge_list2 = list(edge_list.itertuples(index=False))

G1A.add_weighted_edges_from(edge_list2)

print("Total number of nodes: ", int(G1A.number_of_nodes()))
print("Total number of edges: ", int(G1A.number_of_edges()))


# Create dictionary of edge labels: 

edge_labels_key = df_edges[['species','prey']].itertuples(index=False)
edge_labels_dict = dict(zip(edge_labels_key, df_edges['prey_weight_g'].round(decimals = 4)))


# Style and show graph: 

pos = nx.drawing.layout.bipartite_layout(G1A, ['white perch', 'yellow perch']) 

# Make graph styling 
plt.figure(3,figsize=(12,12))
nx.draw(G1A, pos, edge_color='black', width=1, linewidths=1,
    node_size=250,  alpha=0.9, with_labels = True)
nx.draw_networkx_edge_labels(G1A, pos, edge_labels = edge_labels_dict)
plt_size = (105,150)
label_offset = -20
plt.show()
```
