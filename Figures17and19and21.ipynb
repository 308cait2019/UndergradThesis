```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("wc1A2014.csv")
df_white14 = df[df['species']== 'white perch'] #changed df_white

print("Graph 1.a.1: white perch in 2014")

#df.iloc[0:344]

G1A = nx.Graph()

G1A.add_nodes_from(df_white14['prey'])
G1A.add_nodes_from(df_white14['species'])

#edge_list = df.loc[0::346, 'prey_weight_g']

# Here the changes begin:


# Construct the edge_list and add it to graph with weights:
edge_list = df_white14[['species','prey','prey_weight_g']] #changed syntax
edge_list2 = list(edge_list.itertuples(index=False))

G1A.add_weighted_edges_from(edge_list2)

print("Total number of nodes: ", int(G1A.number_of_nodes()))
print("Total number of edges: ", int(G1A.number_of_edges()))


# Create dictionary of edge labels: 

edge_labels_key = df_white14[['species','prey']].itertuples(index=False)
edge_labels_dict = dict(zip(edge_labels_key, df_white14['prey_weight_g'].round(decimals = 4)))


# Style and show graph: 

pos = nx.planar_layout(G1A,scale=1)

# Make graph styling 
plt.figure(3,figsize=(12,12))
nx.draw(G1A, pos, edge_color='black', width=1, linewidths=1,
    node_size=250,  alpha=0.9, with_labels = True)
nx.draw_networkx_edge_labels(G1A, pos, edge_labels = edge_labels_dict)
plt_size = (105,150)
label_offset = -20
plt.show()
```

```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from networkx.algorithms import bipartite 
    
T1 = pd.read_csv("wc1A2014.csv")

print("Graph 1.a.3: both perch species in 2014")

#df.iloc[238:522]

G1A = nx.Graph()

G1A.add_nodes_from(T1['prey'], bipartite = 1)
G1A.add_nodes_from(T1['species'], bipartite = 0) 

#edge_list = df.loc[0::523, 'prey_weight_g']

# Here the changes begin:


# Construct the edge_list and add it to graph with weights:
edge_list = T1.loc[0:523, 'species':'prey_weight_g']
edge_list2 = list(edge_list.itertuples(index=False))

G1A.add_weighted_edges_from(edge_list2)

print("Total number of nodes: ", int(G1A.number_of_nodes()))
print("Total number of edges: ", int(G1A.number_of_edges()))


# Create dictionary of edge labels: 

edge_labels_key = list(T1.loc[0:523, 'species':'prey'].itertuples(index=False))
edge_labels_dict = dict(zip(edge_labels_key, T1['prey_weight_g'].round(decimals = 4)))


# Style and show graph: 

pos = nx.drawing.layout.bipartite_layout(G1A, ['white perch', 'yellow perch']) 

# Make graph styling 
plt.figure(3,figsize=(12,12))
nx.draw(G1A, pos, edge_color='black', width=1, linewidths=1,
    node_size=250,  alpha=0.9, with_labels = True)
nx.draw_networkx_edge_labels(G1A, pos, edge_labels = edge_labels_dict)
plt_size = (105,150)
label_offset = -20
plt.show()
```
